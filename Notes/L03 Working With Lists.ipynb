## üîÅ Concept: The Playlist Loop
You have a playlist of 500 songs. You don't want to press "Play" 500 times.
You want to say: "For every song in the playlist, play it."

This is a **Loop**.

```python
playlist = ['Bohemian Rhapsody', 'Hotel California', 'Imagine', 'Billie Jean']

# Read this as: "For every song IN the playlist..."
for song in playlist:
    print(f"Now playing: {song}")
    print("Great tune!\n")

print("Playlist finished.")
```

### üî¢ Making Lists of Numbers
Sometimes you need a range of numbers, like counting reps at the gym or pages in a book.

```python
# range(1, 5) stops BEFORE 5. It gives 1, 2, 3, 4.
for day in range(1, 6):
    print(f"Day {day} of the challenge complete.")
```

## ‚úÇÔ∏è Slicing a List
Maybe you only want to see the Top 3 movies, or the last 2 emails.
We call this **Slicing**.

```python
top_movies = ['The Godfather', 'Pulp Fiction', 'Inception', 'The Matrix', 'Parasite']

print("Top 3 Movies:")
print(top_movies[0:3]) # Start at 0, stop BEFORE 3

print("Last 2 Movies:")
print(top_movies[-2:])
```

### ¬© Copying a List
If you want to make a specialized playlist based on an old one, you need to copy it.

```python
my_foods = ['Pizza', 'Falafel', 'Carrot Cake']
friend_foods = my_foods[:] # [:] Selects EVERYTHING, creating a copy

my_foods.append('Cannoli')
friend_foods.append('Ice Cream')

print(f"Me: {my_foods}")
print(f"Friend: {friend_foods}")
```
