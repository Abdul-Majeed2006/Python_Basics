# Phase 9: List Comprehensions ⚡

> **Goal:** Express transformations clearly.

## 1. Syntax

`[expression for item in iterable if condition]`

It reads like English.

```python
numbers = [1, 2, 3, 4, 5, 6]

# Traditional
squares = []
for n in numbers:
    if n % 2 == 0:
        squares.append(n**2)

# Comprehension
squares_comp = [n**2 for n in numbers if n % 2 == 0]

print(squares_comp)
```

## 2. The Readability Limit

**Rule:** If a comprehension spans more than 2 lines or has nested loops, write a normal loop.
One-liners are not "pro" if nobody can read them.

```python
# UNREADABLE ❌
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [x for row in matrix for x in row if x % 2 == 0]

# BETTER ✅
flat_better = []
for row in matrix:
    for x in row:
        if x % 2 == 0:
            flat_better.append(x)
```
