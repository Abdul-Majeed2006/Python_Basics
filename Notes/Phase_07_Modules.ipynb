## 1. The `if __name__ == "__main__":` Guard

This prevents your script from executing code when it is imported by another script.

```python
def heavy_computation():
    print("Computing...")

if __name__ == "__main__":
    # This only runs if you run 'python Phase_07.py' directly.
    # It DOES NOT run if you 'import Phase_07'
    heavy_computation()
```

## 2. Circular Imports

If Module A imports Module B, and Module B imports Module A, Python will crash or recurse infinitely.
**Fix:** Refactor shared logic into a third Module C.

## 3. The `__init__.py` file

This file turns a directory into a Package. 
It can be empty, or it can expose specific functions to simplify imports.
