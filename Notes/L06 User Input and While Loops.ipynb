## üó£Ô∏è Concept: User Feedback
Programs are useless if they can't take input.
We use `input()` to ask questions.

```python
# The program pauses here until the user types something and hits ENTER
name = input("Name for the reservation? ")

print(f"Table confirmed for {name}.")
```

### ‚ö†Ô∏è Handling Numbers
`input()` ALWAYS returns a **String** (text). If you want to do math (like splitting a bill), you must convert it.

```python
bill = input("Total Bill Amount: $")
people = input("Split how many ways? ")

# Must convert text to floats/ints!
total = float(bill)
count = int(people)

per_person = total / count

print(f"Each person owes: ${per_person}")
```

## üîÑ While Loops
A `for` loop runs for a set list.
A `while` loop runs **until you tell it to stop**.

Perfect for entering data until you are done.

```python
# Interest Calculator
savings = 1000
year = 1

# Calculate untill we double our money
while savings < 2000:
    savings = savings * 1.05 # 5% Interest
    print(f"Year {year}: ${savings:.2f}")
    year += 1
```

### üè≥Ô∏è The Quit Condition
Letting the user control when to stop.

```python
active = True
guests = []

while active:
    name = input("Enter guest name (or 'quit' to finish): ")
    
    if name == 'quit':
        active = False # Stop the loop
    else:
        print(f"Added {name} to the list.")
        guests.append(name)

print(f"Final Guest List: {guests}")
```
