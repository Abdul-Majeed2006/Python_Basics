## üó£Ô∏è Concept: User Feedback
Programs are useful because they can take instructions from users at runtime.

````python
# The program pauses here until the user types and hits ENTER
name = input("Name for the reservation? ")

print(f"Table confirmed for {name}.")
````

üîß **Engineering Note: Blocking Calls**
`input()` is a **blocking call**. This means your program completely stops and waits forever for the user. In high-performance systems or games, we often use "non-blocking" input so the game keeps running while waiting for a key press.

### ‚ö†Ô∏è Handling Numbers
`input()` ALWAYS returns a **String**. To do math, you must cast it to an `int` or `float`.

````python
bill = input("Total Bill Amount: $")
people = input("Split how many ways? ")

# Must convert text to numeric types
total = float(bill)
count = int(people)

print(f"Each person owes: ${total / count:.2f}")
````

## üîÑ While Loops
While a `for` loop runs over a collection, a `while` loop runs **until a condition becomes False**.

````python
# Savings Goal Tracker
savings = 1000
year = 1

while savings < 2000:
    savings *= 1.05 # 5% compounded annually
    print(f"Year {year}: ${savings:.2f}")
    year += 1
````

‚ö†Ô∏è **Common Mistake: The Infinite Loop**
If your loop condition never becomes False, the program will run forever and freeze your computer.

```python
x = 1
while x < 5:
    print(x) # ‚ùå ERROR: x never changes, loop never ends!
```

üí° **Pro Tip**: Always ensure you have an **update step** (like `x += 1`) or a `break` statement inside your while loops.

### üèÅ The Quit Condition (Using Flags)
Flags are boolean variables that control the overall state of the loop.

````python
active = True
guests = []

while active:
    name = input("Enter guest name (or 'quit' to finish): ")
    
    if name.lower() == 'quit':
        active = False 
    else:
        print(f"Added {name} to list.")
        guests.append(name.title())

print(f"Final Guest List: {guests}")
````

````python
# TODO: Implement the Mission below
total_spent = 0

while True:
    amount = float(input("Enter expense amount (or 0 to stop): "))
    
    if amount == 0:
        break
    elif amount < 0:
        print("Error: Invalid amount.")
        continue
        
    total_spent += amount

print(f"Total Spent: ${total_spent:.2f}")
````
