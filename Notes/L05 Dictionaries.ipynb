## üìñ Concept: Key-Value Pairs
A **List** finds things by *Order* (0, 1, 2).
A **Dictionary** finds things by *Name* (Key).

Think of a **Cocktail Menu** or a **Contact List**.

*   **Key**: "Margarita"
*   **Value**: "Tequila, Lime, Triple Sec"

```python
# Creating a dictionary
# Keys must be unique.

user_profile = {
    'username': 'jdoe88',
    'email': 'jane.doe@example.com',
    'location': 'Chicago',
    'premium_member': True
}

print(user_profile)

# Accessing values by looking up the Key
print(f"User Email: {user_profile['email']}")
```

### üõ†Ô∏è Modifying Dictionaries
You can add new keys or change existing ones. It's dynamic.

```python
cocktail = {'name': 'Old Fashioned', 'price': 12}

# 1. Adding a new item
cocktail['ingredients'] = ['Bourbon', 'Bitters', 'Sugar']
print(f"Added Ingredients: {cocktail}")

# 2. Modifying a value
cocktail['price'] = 14 # Inflation!
print(f"New Price: {cocktail}")

# 3. Removing an item
del cocktail['price']
print(f"Deleted Price: {cocktail}")
```

### üõ°Ô∏è The `.get()` Method (Safety First)
If you try to find a contact that doesn't exist, Python crashes.
Use `.get()` to look safely.

```python
profile = {'name': 'Alex'}

# print(profile['phone']) <-- ERROR! No phone number.

# Safe way:
phone = profile.get('phone', 'No Phone Listed')
print(f"Phone: {phone}")
```

## üîÑ Looping Through Dictionaries
You can loop through just keys, just values, or both.

```python
poll_results = {
    'Alice': 'Python',
    'Bob': 'C',
    'Charlie': 'Java'
}

# Loop Key-Value Pairs
for name, lang in poll_results.items():
    print(f"{name} voted for {lang}.")

print("----")

# Loop Keys only
for name in poll_results.keys():
    print(name)
```

## üê£ Nesting
Putting lists inside dictionaries.

```python
# Dictionary of Lists (A Pizza Order)
# One Key -> Many toppings
order = {
    'crust': 'thick',
    'toppings': ['mushrooms', 'cheese', 'peppers']
}

print(f"You ordered a {order['crust']} crust pizza with:")
for topping in order['toppings']:
    print(f"\t- {topping}")
```
