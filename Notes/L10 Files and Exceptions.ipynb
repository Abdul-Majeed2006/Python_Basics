## üíæ Files: The Travel Journal
Your program's memory is temporary. If you want to keep data (like a travel journal or expense log), you must save it to a file.

*   `'w'`: Write (Starts fresh/erases old data)
*   `'a'`: Append (Adds to the end)
*   `'r'`: Read (Just looking)

```python
# Writing a Journal Entry
filename = 'journal.txt'

with open(filename, 'w') as file_object:
    file_object.write("Day 1: Arrived in Kyoto. It is beautiful.\n")
    file_object.write("Day 2: Ate way too much sushi.\n")
```

```python
# Reading the Journal back
with open(filename, 'r') as file_object:
    content = file_object.read()
    print("--- MY JOURNAL ---")
    print(content)
```

### üì¶ JSON: Saving Settings
JSON is great for saving "App Settings" or structured data, like a user profile.

```python
import json

app_settings = {
    'theme': 'Dark Mode',
    'notifications': True,
    'volume': 80
}

filename = 'settings.json'

# Saving settings
with open(filename, 'w') as f:
    json.dump(app_settings, f)
    print("Settings Saved.")
```

```python
# Loading settings
with open(filename, 'r') as f:
    loaded_settings = json.load(f)
    
print(f"Current Theme: {loaded_settings['theme']}")
```
