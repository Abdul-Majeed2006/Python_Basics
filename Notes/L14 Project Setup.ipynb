## ðŸ› ï¸ Concept: The Workspace
So far, we've only used code that comes with Python.
But the real power is installing *other people's* code (Libraries).

To do this safely, we create a "Virtual Environment" (venv).
Think of it like a separate garage for each car you build, so the parts don't get mixed up.

### 1ï¸âƒ£ Creating a Venv
You run these commands in your **Terminal** (not here in Python).

**Windows:**
`python -m venv .venv`

**Mac/Linux:**
`python3 -m venv .venv`

This creates a hidden folder called `.venv`.

### 2ï¸âƒ£ Activating the Venv
You must turn it on before you start working.

**Windows:**
`.venv\Scripts\activate`

**Mac/Linux:**
`source .venv/bin/activate`

You will see `(.venv)` appear in your terminal.

### 3ï¸âƒ£ Installing Packages (Pip)
Once the venv is active, you can install tools.
For example, `requests` is a famous tool for downloading websites.

`pip install requests`

```python
# Once installed, you can import it like normal
# (This will only work if you have installed 'requests' in your environment!)

import requests

response = requests.get('https://api.github.com')
print(f"Github Status: {response.status_code}")
print("Success! We just downloaded data from the internet.")
```
