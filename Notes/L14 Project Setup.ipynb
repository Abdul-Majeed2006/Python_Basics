## ðŸ› ï¸ Concept: The Workspace
Most software relies on code written by others. To manage these "dependencies" without breaking your computer, we use **Virtual Environments**.

### 1ï¸âƒ£ Creating and Activating a Venv
Run these in your **Terminal** (PowerShell or Bash).

**Step A: Create**
```bash
python -m venv .venv
```

**Step B: Activate**
- **Windows**: `.venv\Scripts\activate` 
- **Mac/Linux**: `source .venv/bin/activate` 

Once active, your terminal will show `(.venv)` before the prompt.

### 2ï¸âƒ£ Installing Packages (`pip`)
Use `pip` to install the tools you need.

````python
# RUN THIS IN TERMINAL: pip install requests

import requests

try:
    response = requests.get('https://api.github.com', timeout=5)
    print(f"Connection Status: {response.status_code} (Success)")
except requests.exceptions.RequestException as e:
    print(f"Connection Failed: {e}")
````

### 3ï¸âƒ£ Locking Dependencies (`requirements.txt`)
Professional projects always include a `requirements.txt` file so other engineers can set up the project exactly like yours.

**To Save your current tools:**
```bash
pip freeze > requirements.txt
```

**To Install a project from scratch:**
```bash
pip install -r requirements.txt
```

ðŸ”§ **Engineering Note: Version Control**
Never commit your `.venv` folder to Git. It is massive and contains system-specific files. Instead, commit your `requirements.txt`. Your collaborators will use it to rebuild their own environment.
