## 1. Stop using `print()` for everything

When your code runs on a server, you can't see the console.
Use **Logging**.

```python
import logging

# Configure once at start of app
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

logging.info("Program started")
logging.warning("Disk space low")
logging.error("Database connection failed")
```

## 2. The Python Debugger (`pdb`)

Put `breakpoint()` in your code to pause execution and inspect variables interactively in the terminal.

## 3. Unit Testing (`unittest`/`pytest`)

Testing proves your code works now, and that future you didn't break it.

```python
import unittest

def add(a, b):
    return a + b

class TestMath(unittest.TestCase):
    def test_add(self):
        self.assertEqual(add(2, 3), 5)
        self.assertEqual(add(-1, 1), 0)

# In notebook we run it differently, but normally 'python -m unittest' works
unittest.main(argv=[''], exit=False)
```
