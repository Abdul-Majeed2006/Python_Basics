## ðŸ·ï¸ Concept: Helping the Computer Help You
Python is "dynamically typed," which means `x` can be a number, then a word, then a sandwich. This is flexible, but dangerous.

Modern Python uses **Type Hints** to tell code editors what kind of data to expect.
This helps catch bugs *before* you run the code.

### ðŸ“ The Syntax
It uses a colon `:`. `variable: type`.

```python
# Without Type Hints (The Old Way)
name = "Alice"
age = 30

# With Type Hints (The Modern Way)
name: str = "Alice"
age: int = 30
height: float = 1.75
is_student: bool = False

# It doesn't change how the code RUNS, but hover over 'name' with your mouse!
print(f"{name} is {age}.")
```

### âž¡ï¸ Function Hints
This is where it is most useful. We use `->` to show what the function gives back.

```python
# This function expects two integers and promises to return an integer
def add_numbers(a: int, b: int) -> int:
    return a + b

result = add_numbers(10, 5)
print(result)
```

### ðŸ“¦ Complex Types (Lists and Dicts)
What if you have a list of numbers?

```python
from typing import List, Dict

# A list that ONLY contains strings
shopping_list: List[str] = ["Apples", "Bananas", "Coffee"]

# A dictionary mapping names (str) to scores (int)
scores: Dict[str, int] = {
    "Alice": 95,
    "Bob": 80
}

print(shopping_list)
```
