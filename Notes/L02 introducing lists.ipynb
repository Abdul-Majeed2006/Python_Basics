## üìã Concept: The Bucket List
A **List** is a collection of items in a specific order.
Think of it like a **Bucket List** or a **Shopping List**.
You can add items, cross them off, or change their order.

```python
# Creating a list
# Lists use square brackets []
bucket_list = ['Japan', 'Iceland', 'New Zealand', 'Peru']

print(bucket_list)
```

### üìç Accessing Items
Lists are ordered. The first item is at position **0** (computers start counting at 0!).

```python
print(bucket_list[0])  # First item
print(bucket_list[1])  # Second item
print(bucket_list[-1]) # Last item (Super useful!)

# Using individual items in sentences
msg = f"My next flight is to {bucket_list[0].upper()}."
print(msg)
```

## ‚úèÔ∏è Modifying the List
Plans change. You might want to swap destinations or add new ones.

```python
# 1. Changing an item
bucket_list[3] = 'Argentina' # Swapping Peru for Argentina
print(bucket_list)

# 2. Adding to the end (Append)
bucket_list.append('Italy')
print(f"Added Italy: {bucket_list}")

# 3. Inserting in the middle
bucket_list.insert(0, 'Thailand') # New priority #1!
print(f"Priority Update: {bucket_list}")
```

### ‚ùå Removing Items
Trip cancelled? Already visited? Time to remove it.

```python
# Using 'del' (Delete by position)
del bucket_list[0] 
print(f"Removed First: {bucket_list}")

# Using 'pop' (Remove last item and use it)
visited_country = bucket_list.pop()
print(f"I just visited {visited_country}!")
print(f"Remaining: {bucket_list}")

# Using 'remove' (Delete by value)
bucket_list.remove('Iceland')
print(f"Too cold: {bucket_list}")
```
