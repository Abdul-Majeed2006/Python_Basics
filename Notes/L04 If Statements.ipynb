## üö¶ Concept: Flow Control
Logic is the "branching" of software: "If it rains, take a taxi. Else, walk."

````python
# The Bouncer Logic
age = 20

if age >= 21:
    print("Welcome to the club! üç∏")
else:
    print("Sorry, come back when you're older.")
````

‚ö†Ô∏è **Common Mistake: Assignment (=) vs Equality (==)**
Using a single `=` inside an `if` statement is a common error (though Python's syntax usually catches this, unlike C).

```python
if age = 21:  # ‚ùå ERROR: Assignment not allowed here
if age == 21: # ‚úÖ Correct: Checking equality
```

### üõ£Ô∏è Multiple Options (elif)
Use `elif` (else if) to check multiple conditions sequentially.

````python
genre = "Sci-Fi"

if genre == "Comedy":
    print("Recommended: 'The Office'")
elif genre == "Sci-Fi":
    print("Recommended: 'Interstellar'")
elif genre == "Horror":
    print("Recommended: 'The Conjuring'")
else:
    print("Recommended: 'Planet Earth'")
````

üí° **Pro Tip: Pythonic Truthiness**
In Python, empty collections (lists, strings, dicts) and the number 0 are considered **False**. Everything else is **True**.

```python
items = []
if items:
    print("List has items")
else:
    print("List is empty") # ‚úÖ This runs
```

## ü§ù Logical Operators (and / or / not)
Combine multiple facts into a single decision.

````python
has_subscription = True
internet_connected = False

# AND: Both must be True
if has_subscription and internet_connected:
    print("Streaming Movie... üçø")
else:
    print("Check internet connection.")

# OR: Only one needs to be True
day = "Saturday"
if day == "Saturday" or day == "Sunday":
    print("Weekend mode activated! üí§")
````

### ‚úÖ Checking Membership (`in`)
Check if an item exists within a collection.

````python
banned_users = ['troll123', 'spambot', 'hacker99']
user = 'alice_nyc'

if user not in banned_users:
    print(f"Welcome back, {user}!")
else:
    print("üö® Security Alert: Access Denied.")
````

üîß **Engineering Note: Entry Points**
You will often see `if __name__ == "__main__":`. This checks if the script is being run directly or imported as a module from another file. It's a standard engineering practice to keep scripts modular.
