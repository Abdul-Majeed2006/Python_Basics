## üö¶ Concept: Flow Control
Life is full of decisions: "If it rains, I'll take a taxi. Else, I'll walk."
Python uses `if` statements to handle this logic.

```python
# The Bouncer Logic
age = 20

if age >= 21:
    print("Welcome to the club! üç∏")
else:
    print("Sorry, come back when you're older.")
```

### üõ£Ô∏è Multiple Options (elif)
What if there are more than two choices? Like a movie recommender.

```python
genre = "Sci-Fi"

print(f"User likes: {genre}")

if genre == "Comedy":
    print("Recommended: 'The Office'")
elif genre == "Sci-Fi":
    print("Recommended: 'Interstellar'")
elif genre == "Horror":
    print("Recommended: 'The Conjuring'")
else:
    print("Recommended: 'Planet Earth'")
```

## ü§ù Logical Operators (and / or)
Sometimes you need to check two things at once.

```python
has_subscription = True
internet_connected = False

# AND: Both must be true
if has_subscription and internet_connected:
    print("Streaming Movie... üçø")
else:
    print("Error: Check account or internet.")

# OR: At least one must be true
day = "Saturday"
if day == "Saturday" or day == "Sunday":
    print("It's the weekend! Sleep in.")
```

### ‚úÖ Checking a List
Is an item in the list?

```python
banned_users = ['troll123', 'spambot', 'hacker99']
user = 'alice_nyc'

if user not in banned_users:
    print(f"Access granted for {user}.")
else:
    print("Account suspended.")
```
