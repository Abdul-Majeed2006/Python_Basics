How do you know your code works? You run it.
But what if you change something later? How do you know you didn't break it?

You write **Tests**.
Tests are code that checks your code.

### üß™ The `unittest` Module
Python comes with a built-in testing tool called `unittest`.

```python
import unittest

# Here is a logical function: splitting a bar tab
def split_tab(total, people):
    if people <= 0:
        return 0
    return total / people

# Here is the Test Suite to prove it works
class TestTab(unittest.TestCase):

    def test_even_split(self):
        # $100 split by 2 people should be $50
        result = split_tab(100, 2)
        self.assertEqual(result, 50)
        
    def test_solo_drinker(self):
        # $30 tab for 1 person should be $30
        self.assertEqual(split_tab(30, 1), 30)

# This line runs the tests when you click Play
unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

## üö¶ Assertions

*   `assertEqual(a, b)`: Checks if a == b
*   `assertTrue(x)`: Checks if x is True
*   `assertIn(item, list)`: Checks if item is in the list
