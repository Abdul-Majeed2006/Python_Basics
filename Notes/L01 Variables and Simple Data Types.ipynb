## ðŸ“¦ Concept: Variables are Labels
Imagine you are moving houses or packing for a big trip.
A **variable** is just a focused label on a box.

*   Box Content: A pair of hiking boots.
*   Variable Name: `footwear`

You don't need to know exactly *how* the box is made, you just need to know what's inside.

```python
# Creating a variable (Assigning a value)
destination = "Paris"
days_remaining = 10

print(destination)
print(days_remaining)
```

## ðŸ§µ Strings (Text)
Strings are just pieces of text. You can use single `' '` or double `" "` quotes.

```python
city = "new york city"

# Methods: Tools to change the text
print(city.title()) # Capitalize Each Word
print(city.upper()) # YELL IT
print(city.lower()) # whisper it
```

### f-strings (The Glue)
The best way to mix variables into sentences.

```python
first_name = "James"
last_name = "Bond"

# Look for the 'f' before the quotes
full_intro = f"The name is {last_name}. {first_name} {last_name}."
print(full_intro)
```

## ðŸ”¢ Numbers
*   **Integers (int)**: Whole numbers. (Years, Count of people)
*   **Floats (float)**: Decimals. (Money, Distance)

```python
# Integers
salary = 85000
bonus = 5000
total_comp = salary + bonus
print(f"Total Comp: ${total_comp}")

# Floats
bill = 45.50
tip = bill * 0.20
print(f"Tip Amount: ${tip}")
```

### ðŸ’¡ Underscores in Numbers
You can use underscores to make big numbers readable. Python ignores them.

```python
views = 1_000_000_000
print(views)
```
