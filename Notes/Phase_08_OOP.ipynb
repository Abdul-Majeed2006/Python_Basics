OOP is about **Encapsulation** (hiding details) and **Abstraction** (simplifying interfaces).

## 1. Composition > Inheritance

**Inheitance:** "Is-a" relationship. (Dog is a Animal)
**Composition:** "Has-a" relationship. (Car has a Engine)

**Constraint:** Favor Composition. Inheritance hierarchies get brittle fast.

```python
# Composition Example
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()  # Use the Engine, don't inherit from it

    def drive(self):
        self.engine.start()
        print("Car moving")
```

## 2. Dataclasses (Modern Python)

Stop writing boilerplate `__init__` and `__repr__` methods for data holders.

```python
from dataclasses import dataclass

@dataclass
class Point:
    x: float
    y: float
    name: str = "Point"

p1 = Point(10.5, 20.0)
print(p1)  # Automatically prints nice string!
```

## 3. Class vs Instance Variables

A class variable is shared by ALL instances. An instance variable belongs to ONE.

```python
class Dog:
    kind = "Canine" # Class variable (Shared)

    def __init__(self, name):
        self.name = name # Instance variable (Unique)

d1 = Dog("Buddy")
d2 = Dog("Rex")

Dog.kind = "Wolf" # Changed for ALL dogs unexpectedly
print(d1.kind) # Wolf
```
