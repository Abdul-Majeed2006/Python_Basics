## ‚ö° Concept: The Speed Lane
You already know how to make lists using a loop. But professional Python developers rarely write simple loops to make lists.
They use **List Comprehensions**.

It's a way to compress a 3-line loop into **1 line**.

### üê¢ The Old Way (For Loop)
Let's say we have a list of luggage weights in Kg, and we want to know them in Lbs.

```python
weights_kg = [20, 15, 23, 10]
weights_lbs = []

for w in weights_kg:
    weights_lbs.append(w * 2.2)

print(weights_lbs)
```

### üêá The New Way (List Comprehension)
We can do the same thing in one line.
Syntax: `[ expression for item in list ]`

```python
# Read this as: "Give me w * 2.2 FOR every w IN weights_kg."
weights_lbs = [w * 2.2 for w in weights_kg]

print(weights_lbs)
```

### üïµÔ∏è Filtering with If
You can also add an `if` statement to the end to filter items.

```python
flight_costs = [150, 400, 90, 850, 120]

# Find only the cheap flights (under $200)
cheap_flights = [cost for cost in flight_costs if cost < 200]

print(f"All Flights: {flight_costs}")
print(f"Steals: {cheap_flights}")
```

```python
# Real World Example: Processing User Input
usernames = ['  john  ', '  SARAH ', 'mike']

# Clean up all strings at once (strip whitespace + lower case)
cleaned_users = [name.strip().lower() for name in usernames]

print(cleaned_users)
```
