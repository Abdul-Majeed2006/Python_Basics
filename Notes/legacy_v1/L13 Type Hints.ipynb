## ðŸ·ï¸ Concept: Helping the Computer Help You
Python is "dynamically typed," which means a variable can change its type at any time. This flexibility is great for small scripts but a nightmare for large software systems.

**Type Hints** allow you to document exactly what kind of data should be in a variable.

```python
# The Old Way (Ambiguous)
user = "Alice"
age = 25

# The Modern Way (Explicit)
user_name: str = "Alice"
user_age: int = 25
account_balance: float = 1250.50
is_premium: bool = True
```

ðŸ”§ **Engineering Note: Static Analysis**
Type hints don't actually change how Python runs the code. However, tools like **Mypy**, **Pyright**, or your IDE (VS Code/PyCharm) use these hints to warn you of potential bugs *before* you run the program. This is called "Static Analysis."

### âž¡ï¸ Function Hints
Explicitly stating what a function expects and what it produces.

```python
def calculate_area(radius: float) -> float:
    """Calculates area and promises to return a float."""
    import math
    return math.pi * (radius ** 2)

area: float = calculate_area(5.5)
print(f"Area: {area:.2f}")
```

### ðŸ“¦ Complex Structures
For lists, dictionaries, and optional values, we use the `typing` module.

```python
from typing import List, Dict, Optional

# A list that must only contain integers
scores: List[int] = [95, 88, 72]

# A dictionary mapping strings to integers
inventory: Dict[str, int] = {"Apples": 10, "Bananas": 5}

# Optional: Can be a string OR None
middle_name: Optional[str] = None

print(f"Inventory Keys: {list(inventory.keys())}")
```

1. Define a function `greet_user`.
2. Add type hints: 
   - `name` should be a `str`.
   - `repeat_count` should be an `int` (default value of 1).
   - The return value should be a `List[str]` (a list of greeting strings).
 3. Call the function and print the result.

```python
# TODO: Implement the Mission below
from typing import List

def greet_user(name: str, repeat_count: int = 1) -> List[str]:
    greetings: List[str] = []
    for _ in range(repeat_count):
        greetings.append(f"Hello, {name}!")
    return greetings

results: List[str] = greet_user("Zakari", 3)
for g in results:
    print(g)
```
